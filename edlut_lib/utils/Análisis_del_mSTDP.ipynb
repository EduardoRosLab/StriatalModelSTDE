# Análisis de la mSTDP

## Wolfram:

### Petición

```
c'(t)=-c(t)/x; n'(t)=-n(t)/y; w'(t)=c(t)*(n(t) - b); c(0)=u, n(0)=v, w(0)=z
```

### Resultado

```
c(t) = u e^(-t/x)
```

```
n(t) = v e^(-t/y)
```

```
w(t) = -(exp(-(t*(tau_c+tau_n))/(tau_c*tau_n)-t/tau_c)*(-b*c_0*tau_c**2*exp((t*(tau_c+tau_n))/(tau_c*tau_n))+b*c_0*tau_c**2*exp((t*(tau_c+tau_n))/(tau_c*tau_n)+t/tau_c)-b*c_0*tau_c*tau_n*exp((t*(tau_c+tau_n))/(tau_c*tau_n))+b*c_0*tau_c*tau_n*exp((t*(tau_c+tau_n))/(tau_c*tau_n)+t/tau_c)+c_0*n_0*tau_c*tau_n*exp(t/tau_c)-c_0*n_0*tau_c*tau_n*exp((t*(tau_c+tau_n))/(tau_c*tau_n)+t/tau_c)-tau_c*w_0*exp((t*(tau_c+tau_n))/(tau_c*tau_n)+t/tau_c)-tau_n*w_0*exp((t*(tau_c+tau_n))/(tau_c*tau_n)+t/tau_c)))/(tau_c+tau_n)
```

### Coste computacional de w(t):

- 10 exponenciales
- 18 divisiones
- 47 multiplicaciones
- 25 sumas/restas/cambios de signo


## Ecuaciones:

### Derivadas del trazo de eligibilidad, dopamina y peso sináptico

$c'(t) = -\frac{c(t)}{\tau_c} \\
n'(t) = -\frac{n(t)}{\tau_n} \\
w'(t) = c(t)\cdot \left(n(t)-b\right )\\
$

### Evolución del trazo de eligibilidad $c$ y dopamina $n$

$c(t) = c_0 \cdot e^{-\Delta t/\tau_c}\\
n(t) = n_0 \cdot e^{-\Delta t/\tau_n}$

### Evolución del peso sinaptico $w$

Variables iniciales: 

$\tau_c, \tau_n, \Delta t, b, c_0, n_0 $

Variables intermedias y resultado:

$\tau_{c^2} = \tau_c^2 \\
\tau_{n^2} = \tau_n^2 \\
\tau_{cn} = \tau_c \cdot \tau_n \\
u_0 = \frac{\Delta t \cdot (\tau_c + \tau_n)}{\tau_{cn}} \\
u_2 = e^{u_0} \\
u_4 = e^{\Delta t/\tau_c} \\
u_3 = u_2 \cdot u_4 \\
u_1 = \frac{1}{u_3} \\
a_1 = b \cdot c_0 \cdot (\tau_{c^2}+\tau_{cn}) \cdot (u_3 - u_2) \\
a_2 = c_0 \cdot n_0 \cdot \tau_{cn} \cdot (u_4 - u_3) \\
a_3 = w_0 \cdot u_3 (\tau_c + \tau_n) \\
w(t) = -\frac{u_1 \cdot \left (a_1 + a_2 - a_3 \right )}{\tau_c + \tau_n}$

### Coste computacional final:

- 2 exponenciales
- 4 divisiones
- 14 multiplicaciones
- 9 sumas/restas/cambios de signo

# Algoritmo

En las ecuaciones no se indica que el peso sináptico debe ser positivo, pero si en el paper. En una implementación time-driven no sería problema simplemente acotar el peso después de los cálculos, pero no es posible hacerlo en una implementación event-driven debido a un caso donde hay problemas.

Ocurre cuando $n(t)-b$ cambia de signo durante el intervalo $t\in[0,\Delta t]$. Esto sólo da problemas cuando $n(0)>b$, $n(\Delta t)<b$ y $c(t)<0$ con $t\in [0, \Delta t]$. En esta situación $w(0)$ y $w(\Delta t)$ son positivas, pero puede haber un intervalo $t \in (t_0, t_1)$ en el que $w(t) < 0$.

````python
def corner_case_example():
    b = 0.5
    t = np.linspace(0.0,1.0)
    tau_n = 0.2
    tau_c = 1.0
    learning = 0.3
    n = 1.5 * np.exp(-t/tau_n)
    c = -1 * np.exp(-t/tau_c)
    w_event = np.zeros_like(n)
    w_event[0] = 0.4
    w_time = w_event.copy()
    for i in range(1, w_time.shape[0]):
        w_event[i] = w_event[i-1] + c[i]*(n[i]-b)*learning
        w_time[i] = w_time[i-1] + c[i]*(n[i]-b)*learning
        w_time[i] = max(0, w_time[i])

    plt.plot(t, n, 'r', label='n(t)')
    plt.plot(t, c, 'b', label='c(t)')
    plt.plot(t, w_event, 'g--', label='w_event')
    plt.plot(t, w_time, 'g-', label='w_time')
        
    plt.hlines(0, 0, 1, colors='black', linestyles='dotted')
    plt.hlines(b, 0, 1, colors='red', linestyles='dotted', label='b')
    
    plt.legend()
corner_case_example()
````

Ese es el único caso que hay que tratar ya que:

- Si $c(t)>0$ entonces $w(\Delta t)<0$, y es fácil acotar el valor de $w$ al final del aprendizaje.
- Si $n(0)<b$ entonces $n(t)-b$ no cambia de signo en todo el intervalo.

La forma de tratarlo es la siguiente:

1. Comprobamos que $n(0)>b$, $n(\Delta t)<b$ y $c(t)<0$. Si no es así, calculamos $w(\Delta t)$ normalmente. 
2. Si sí, buscamos cuándo $n(t) = b$ en el intervalo $t\in[0, \Delta t]$. Es decir, comprobamos si existe $t = -\tau_n\cdot \log\frac{b}{n_0}$:
    1. Si $t$ no existe o no esta en el rango $[0, \Delta t]$, calculamos $w(\Delta t)$ normalmente.
    2. Si $t$ existe y esta en el rango $[0, \Delta t]$, entonces 
        1. Calculamos el valor de $w(t)$
            1. Si $w(t)\ge 0$, calculamos $w(\Delta t)$ normalmente
            2. Si $w(t)<0$, entonces 
                1. $w_0\mapsto 0$
                2. $c_0 \mapsto c(t)$
                3. $n_0 \mapsto n(t)$
                4. $\Delta t \mapsto \Delta t - t$
                5. Calculamos $w(\Delta t)$ con los nuevos valores


# Implementación

## Naïf

````python
def event_driven_weight_change_naif(dt, tau_c, tau_n, b, c_0, n_0, w_0):
    tau_c2 = tau_c * tau_c
    tau_n2 = tau_n * tau_n
    tau_cn = tau_c * tau_n

    u0 = (dt*(tau_c + tau_n)) / tau_cn

    u2 = exp(u0)           # u2 = exp(u0)
    u4 = exp(dt/tau_c)     # u4 = exp(dt/tau_c)
    u3 = u2 * u4           # u3 = exp(u1 + dt/tau_c)
    u1 = 1.0 / u3          # u1 = exp(-u1 - dt/tau_c)

    a1 = b*c_0 * (tau_c2 + tau_cn)*(u3 - u2)
    a2 = c_0*n_0*tau_cn * (u4 - u3)
    a3 = w_0*u3 * (tau_c + tau_n)
    
    new_w = -(u1 * (a1 + a2 - a3)) / (tau_c + tau_n)

    return max(0.0, new_w)
#     return new_w
````

## Corregida 

````python
def event_driven_weight_change(dt, tau_c, tau_n, b, c_0, n_0, w_0):

    n_dt = n_0 * exp(-dt/tau_n)
    if n_0>b and n_dt<b and c_0<0:
        t = -tau_n * np.log(b/n_0)  # t∊[0,Δt] : n(t)=b
        if t<0.0 or t>dt:
            w_dt = event_driven_weight_change_naif(dt, tau_c, tau_n, b, c_0, n_0, w_0)
        else:
            w_t = event_driven_weight_change_naif(t, tau_c, tau_n, b, c_0, n_0, w_0)
            if w_t > 0.0:
                w_dt = event_driven_weight_change_naif(dt, tau_c, tau_n, b, c_0, n_0, w_0)
            else:
                w_0 = 0.0
                c_0 = c_0 * exp(-t/tau_c)
                n_0 = n_0 * exp(-t/tau_n)
                dt = dt - t
                w_dt = event_driven_weight_change_naif(dt, tau_c, tau_n, b, c_0, n_0, w_0)
    else:
        w_dt = event_driven_weight_change_naif(dt, tau_c, tau_n, b, c_0, n_0, w_0)
    
    return w_dt
````

# Ejemplo de uso

````python
def event_driven(naif=True):
    global t, w, c, n, tau_c, tau_n, b
    
    step = 4.0
    for i in range(int(4.0/step)):
        c_0 = c[-1]
        n_0 = n[-1]
        w_0 = w[-1]
        t_0 = t[-1]

        t.append( t_0 + step )
        c.append( c_0 * exp(-step/tau_c) )
        n.append( n_0 * exp(-step/tau_n) )    
        if naif:
            w.append(event_driven_weight_change_naif(step, tau_c, tau_n, b, c_0, n_0, w_0))
        else:
            w.append(event_driven_weight_change(step, tau_c, tau_n, b, c_0, n_0, w_0))

def time_driven():
    global t, w, c, n, tau_c, tau_n, b

    step = 1e-3
    for i in range(int(4.0/step)):
        c0 = c[-1]
        n0 = n[-1]
        w0 = w[-1]
        t0 = t[-1]

        t.append( t0 + step )
        c.append( c0 * exp(-step/tau_c) )
        n.append( n0 * exp(-step/tau_n) )    
#         w.append(w0 + step*(c0*(n0-b)))
        w.append(max(0.0, w0 + step*(c0*(n0-b))))

def simulate_both_methods():
    global t, w, c, n, tau_c, tau_n, b    
    
    tau_c = 2.0
    tau_n = 1.0
    b = 0.5

    t = [0.0]
    w = [1.5]
    c = [-3.0]
    n = [2.0]

    time_driven()
    c[-1] = -1.0
    time_driven()
    n[-1] = 3.0
    time_driven()

    plt.rcParams['figure.figsize'] = (10,5)
    plt.plot(t,c, '-', label='c')
    plt.plot(t,n, '-', label='n')
    plt.plot(t,w, '-', label='w')
    plt.legend()

    t = [0.0]
    w = [1.5]
    c = [-3.0]
    n = [2.0]

    event_driven(naif=False)
    c[-1] = -1.0
    event_driven(naif=False)
    n[-1] = 3.0
    event_driven(naif=False)

    plt.rcParams['figure.figsize'] = (10,5)
    plt.plot(t,c, 'o', label='c')
    plt.plot(t,n, 'o', label='n')
    plt.plot(t,w, 'o', label='w')
    plt.legend()
    
    plt.plot([0,12], [0,0], 'r--')
    
simulate_both_methods()
````
